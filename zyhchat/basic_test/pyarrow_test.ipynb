```python
import pandas as pd
import pyarrow as pa
import pyarrow.parquet as pq

# 1. 从 Pandas 创建 Arrow Table
df = pd.DataFrame({"name": ["alice", "bib"], "age": [25, 30]})
table = pa.Table.from_pandas(df)

# 2. 写入 Parquet 文件（高效压缩）
pq.write_table(table, "zeta_data.parquet")

# 3. 读取 Parquet 文件（可只读部分列！）
t2 = pq.read_table("zeta_data.parquet", columns=["name"])
df2 = pa.Table.to_pandas(t2)
print(df2)
```
