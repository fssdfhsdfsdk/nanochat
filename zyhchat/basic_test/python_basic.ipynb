## contextlib

```python
from contextlib import contextmanager
import time

@contextmanager
def timer():
    start = time.time()
    try:
        yield  # 这里相当于 __enter__ 返回后，执行 with 块
    finally:
        print(f"耗时: {time.time() - start:.2f} 秒")

# 使用
with timer():
    time.sleep(1)
```

```python
from contextlib import contextmanager, nullcontext

@contextmanager
def test_log(msg):
    print(f"test log: {msg}")
    try:
        yield
    finally:
        print("log finish")

def books_func(is_log=True):

    log_cx = test_log("books need log.") if is_log else nullcontext()
    with log_cx:
        pass
    print("--------")

books_func()
books_func(is_log=False)
```
